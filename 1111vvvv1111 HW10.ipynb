# **2. 데이터 전처리**

```python
all_data=pd.concat([train['review'], test['review']],ignore_index=True)
all_data
```

```python
review0=pd.concat([train['review'], test['review']],ignore_index=True)
review0
```

# **3. Tokenizer: fit_on_texts**

```python
from keras.preprocessing.text import Tokenizer
tk=Tokenizer()
tk.fit_on_texts(review0)
```

```python
tk.word_index
```

# **4. text_to_sequences**

```python
tseq=tk.texts_to_sequences(review0) 
```
