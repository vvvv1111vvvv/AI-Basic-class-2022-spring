# **1.cifar10 dataset으로 선형모델 학습**

# **(1) 데이터 불러오기: keras.datasets.cifar10**

```python
from tensorflow.keras.datasets.cifar10 import load_data
(train_x, train_y), (test_x,test_y)=load_data()
```

```python
train_x.shape, train_y.shape, test_x.shape, test_y.shape  #이미지사이즈;32x32x3
```

input_shape(32,32,3,)

```python
np.unique(train_y,return_counts=True) 
#lable(train_y)을 unique를 사용해 조사 0~9 10개의 클래스, 분포:각각500개

#10개의 클래스에 다중분류이다
```

# **(2) 데이터 전처리 : /255**

```python
train_x=train_x/255
test_x=test_x/255
```

# **(3) 선형모델 라이브러리 불러오기**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

# **(4) 선형 모델 설정**

```python
lm=Sequential()
lm.add(Flatten(input_shape=(32,32,3,)))
lm.add(Dense(10, activation='softmax'))
```

# **(5) complie 설정**

```python
lm.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics='acc')
```

# **(6) fit (학습 진행) (validation_split=0.1)**

```python
hist=lm.fit(train_x,train_y,epochs=200, batch_size=500, validation_split=0.1)
```

# **(7) loss와 val_loss 그래프 그리기**

```python
loss=hist.history['loss']          #손실
val_loss=hist.history['val_loss']  # validation 에 대한 손실

acc=hist.history['acc']            #정확도
val_acc=hist.history['val_acc']    #validation 에 대한 정확도

ep=np.arange(1, len(loss)+1)   #인덱스

import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.plot(ep,loss,'b', label='loss')
plt.plot(ep,val_loss,'g', label='val_loss')
plt.title('Loss')
plt.legend()
```

# **(8) acc와 val_acc 그래프 그리기**

```python
plt.figure(figsize=(10,8))
plt.plot(ep,acc,'b', label='acc')
plt.plot(ep,val_acc,'g', label='val_acc')
plt.title('Accuracy')
plt.legend()
```

# **(9) test data evaluate**

```python
lm.evaluate(test_x,test_y)
```
