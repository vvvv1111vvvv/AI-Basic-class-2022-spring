```python
#6-2.함수   𝐲  =  −𝟐𝐱𝟐  + 𝟑𝐱의   그래프를  그리시오. (𝒙𝟐: x**2)
```

```python
x=np.linspace(0.1,4.300)
y=-2*x**2+3*x
plt.axvline(x=0,color='k')
plt.axhline(y=0, color='k')
plt.title('Function Graph')
plt.plot(x,y,'b',label='y=-2x^2+3x')
plt.legend()  #도록
plt.show()
```

```python
#6-3.(1) sklearn.datasets의  iris 데이터를  X,y로  load하여   X를   pandas로  변형하여  파일을  만들고,
```

```python
from sklearn.datasets import load_iris
X1,y1=load_iris(return_X_y=True)
iris1=pd.DataFrame(X1)
iris1
```

```python
#6-3(2).그   마지막  열에  y를   추가하시오.
```

```python
iris1['y1']=y1
iris1
```

```python
#6-3(3)2. 1번의  1,2,3,4번째  각   열을   y축으로  하는   그래프를  그리시오. (x축은  index)
```

```python
xaxis=np.array( [int(i) for i in range(150)]) 
epochs=np.arange(1,len(xaxis)+1)
plt.xlim(1,len(xaxis)+1) 
plt.plot(epochs,iris1[0],'r',label='iris[0]') 
plt.plot(epochs,iris1[1],'g',label='iris[1]')
plt.plot(epochs,iris1[2],'b',label='iris[2]')
plt.plot(epochs,iris1[3],'k',label='iris[3]')
plt.title('Accuracy')
plt.legend()
```
